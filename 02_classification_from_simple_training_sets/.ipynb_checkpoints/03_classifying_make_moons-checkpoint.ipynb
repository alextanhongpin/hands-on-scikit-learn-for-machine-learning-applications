```python
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
import pandas as pd
```

```python
X, y = make_moons(n_samples=1000, shuffle=True, 
                  noise=0.2, random_state=0)

df = pd.DataFrame(dict(x=X[:,0], y=X[:, 1], label=y))
colors = {0: 'magenta', 1: 'cyan'}
fig, ax = plt.subplots()
data = df.groupby('label')
for key, label in data:
    label.plot(ax=ax, kind='scatter', x='x', y='y', label=key,
               color=colors[key])
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
def get_scores(model, X_train, X_test, y_train, y_test):
    y_ptrain = model.predict(X_train)
    y_ptest = model.predict(X_test)
    
    acc_train = accuracy_score(y_train, y_ptrain)
    acc_test = accuracy_score(y_test, y_ptest)
    name = model.__class__.__name__
    return (name, acc_train, acc_test)
```

```python
X, y = make_moons(n_samples=1000, shuffle=True,
                  noise=0.2, random_state=0)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
knn = KNeighborsClassifier().fit(X_train, y_train)
accuracy = get_scores(knn, X_train, X_test, y_train, y_test)
print(f'<< {accuracy[0]} >>')
print(f'train: {accuracy[1]} test: {accuracy[2]}')
```

```python
svm = SVC(gamma='scale', random_state=0)
svm.fit(X_train, y_train)
accuracy = get_scores(svm, X_train, X_test, y_train, y_test)
print(f'<< {accuracy[0]} >>')
print(f'train: {accuracy[1]} test: {accuracy[2]}')
```
